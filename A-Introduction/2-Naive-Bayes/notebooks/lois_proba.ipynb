### 1.1. Mesurer la vraisemblance d'une réalisation

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
pdf1 = norm.pdf(x, loc=0, scale=1)
pdf2 = norm.pdf(x, loc=0, scale=0.5)
pdf3 = norm.pdf(x, loc=2, scale=3)
```

```python
fig = plt.figure()
plt.plot(x, pdf1, label=r'$\mu=0, \sigma$=1', color='blue')
plt.plot(x, pdf2, label=r'$\mu=0, \sigma$=0.5', color='red')
plt.plot(x, pdf3, label=r'$\mu=2, \sigma$=3', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Densité de probabilité de la loi normale centrée réduite')
plt.legend()
plt.show()
```

### 1.2. Echantillonner des réalisations : méthode de la transformée inverse

La méthode de la transformée inverse est une méthode pour échantillonner une variable aléatoire dont on connait la fonction de répartition. Elle consiste à échantillonner une variable uniforme sur l'intervalle [0,1], puis à appliquer l'inverse de la fonction de répartition pour obtenir une réalisation de la variable d'intérêt. Pour rappel, la fonction de répartition d'une variable aléatoire $X$ est la fonction $F$ définie ainsi :
$$
\begin{align*}
    F(x) = \mathbb{P}(X \leq x).
\end{align*}
$$
En pratique, pour la loi normale, on peut utiliser la fonction `scipy.stats.norm.cdf` (*cumulative density function* en anglais).

```python
cdf = norm.cdf(x, loc=0, scale=1)
uniform_samples = np.random.rand(10)
```

```python
fig = plt.figure()
plt.plot(x, cdf, color='blue')
plt.scatter(np.zeros_like(uniform_samples), uniform_samples, color='red', marker='x')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Fonction de répartition de la loi normale centrée réduite')
plt.show()
```
